```python
columns_to_check =  ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']


for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder\

label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])

df = pd.get_dummies(df, columns=['CLASS'], drop_first=True)
```

```python
df = df.reset_index()
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
min_max_scaler = MinMaxScaler()

# List of numerical columns to normalize
numerical_cols = ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']

# Apply Min-Max scaling
df[numerical_cols] = min_max_scaler.fit_transform(df[numerical_cols])

# Check the transformed values
print(df[numerical_cols].head())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Check the transformed values
print(df[numerical_cols].head())
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('adult.csv')

# Assuming your DataFrame is named df
df.replace('?', np.nan, inplace=True)
```

```python
numerical_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
```

```python
df = df.dropna(subset=numerical_cols)
```

```python
# Impute missing categorical values with mode
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 
                    'relationship', 'race', 'gender', 'native-country', 'income']
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```
